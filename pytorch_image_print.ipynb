```python
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np



# dataset 객체를 불러올 떄는 데이터를 전처리하는 부분이 들어간다.
transform_train = transforms.Compose([
    transforms.ToTensor(), # 이미지가 자동으로 0과 1 사이의 값으로 변경됨
])

train_dataset = torchvision.datasets.CIFAR10(
    root = './data',
    train = True,
    download = True,
    transform = transform_train
)

train_loader = torch.utils.data.DataLoader(
    train_dataset,
     batch_size = 128,
     shuffle = True,
     num_workers = 4
     )


def custom_imshow(img):
  img = img.numpy()
  plt.imshow(np.transpose(img, ( 1, 2, 0)))
  plt.show()


  # matplotlib는 알아서 정상적인 이미지로 출력함[0-1], opencv에서는 추가적인 전처리가 필요함
  # pytorch: [Batch size, Channel, Width, Height] 순으로 저장 ====> matplotlib: [Width, Height, Channel] 형태로 변경 필요 :: numpy의 transpose()함수 이용

  # imshow() 함수를 이용하여 데이터셋의 이미지를 배치당 하나씩 출력
def process():
  for batch_idx, (inputs, targets) in enumerate(train_loader):
    custom_imshow(inputs[0])

process()
```
