```python
# 새폴더로 만들지 않기.(숨겨진 파일도 같이 불러짐), 디렉토리 만드는 명령어로 하기
import os
import shutil
from bing_image_downloader.bing_image_downloader import downloader 

directory_list = [
    './custom_dataset/train/',
    './custom_dataset/test/'
]

# 데이터셋 저장 디렉토리 생성
for directory in directory_list:
  if not os.path.isdir(directory):
    os.makedirs(directory)
```

```python
# 분리해주는 코드 작성(사이킷런 또는 직접 분류해도 상관 없음)

def dataset_split(query, train_cnt):

  for directory in directory_list:
    if not os.path.isdir(directory + '/' + query):
      os.makedirs(directory + '/' + query)   # query라는 폴더 만들어짐

  cnt=0

  # 데이터 분리
  for file_name in os.listdir(query):
    if cnt < train_cnt:
      print(f'[Train Dataset] {file_name}')
      shutil.move(query + '/' + file_name, './custom_dataset/train/' + query + '/' + file_name)
    else:
      print(f'[Test Dataset] {file_name}')
      shutil.move(query + '/' + file_name, './custom_dataset/test/' + query + '/' + file_name)
    cnt += 1

  # 나눈 데이터 삭제
  shutil.rmtree(query)
```

```python
import torch
import torch.nn
import torch.optim as optim
import torchvision
from torchvision import datasets, models, transforms

import numpy as np
import time

import matplotlib.pyplot as plt
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
# dataset 변형(resize, 변형, 증감 등)
# 데이터셋을 불러올 때 사용할 변형 정의
transforms_train = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],[0.229,0.224,0.225])
])

transforms_test = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],[0.229,0.224,0.225])
])
```

```python
# 정답 데이터까지 같이 있어야함, 자동으로 생성해주는 것이 있음
# 영문 폴더 이름 사용 요망( 폴더 이름으로 라벨링)

train_datasets = datasets.ImageFolder('./custom_dataset/train', transforms_train)
test_datasets = datasets.ImageFolder('./custom_dataset/test', transforms_test)
```

```python
# 배치 단위로 불러오기

train_dataloader = torch.utils.data.DataLoader(
    train_datasets,
    batch_size=4,
    shuffle=True,
    num_workers=2
)
test_dataloader = torch.utils.data.DataLoader(
    test_datasets,
    batch_size=4,
    shuffle=True,
    num_workers=2
)

print('학습데이터셋의 크기 : ', len(train_datasets))
print('테스트데이터셋 크기 : ', len(test_datasets))

class_names = train_datasets.classes
print('클래스: ', class_names)
```

```python
# 데이터를 가져와서 확인
# 이미지를 출력하는 코드(함수로 만들어서 반복문 돌리기)

def imshow(input, title):
  # tensor 형태를 numpy로 변환해야함 torch.Tensor를 numpy 객체로 변환
  print(input.numpy().shape)   #//

  input = input.numpy().transpose((1, 2, 0)) # 배치를 바꿈
  
  # 색에 제대로 나오게
  mean  = np.array([0.485, 0.456, 0.405])
  std = np.array([0.299, 0.224, 0.225])

  input = std * input + mean
  input = np.clip(input, 0, 1)

  print('input shape : ', input.shape)
  print(title)


  plt.imshow(input)
  plt.title(title)
  plt.show()
```

```python
iterator = iter(train_dataloader)

inputs, classes = next(iterator)
print(classes)
print(inputs.shape) # batch size, rgb, width, height 로 만들어짐

imshow(inputs, classes)
```

```python
iterator = iter(train_dataloader)

inputs, classes = next(iterator)
print(classes)
print(inputs.shape) # batch size, rgb, width, height 로 만들어짐
out = torchvision.utils.make_grid(inputs)  # 배치단위의 데이터를 하나로 합쳐주는 mkgrid 함수
print(out.shape)
imshow(out, title=[class_names[x] for x in classes ])
# data준비는 끝남
```

```python
model = models.resnet34(pretrained=True)
model
```

```python
import torch.nn as nn
for param in model.parameters():
  param.requires_grid=False   # Pytorch는 grid 가 기본값이 True, freeze 시키기, 다양하게 검색해보기

num_features = model.fc.in_features
model.fc = nn.Linear(num_features, 3)
model = model.to(device)
```

```python
model
```

```python
for name, module in model.named_parameters():
  print(name + '  :  ' + str(module.requires_grad))
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr = 0.001 , momentum=0.9 )
```

```python
num_epochs = 50
model.train()
start_time = time.time()

for epoch in range(num_epochs):
  running_loss = 0
  running_corrects = 0

  for inputs, labels in train_dataloader:
    inputs = inputs.to(device) # 모델이 cuda에 있기때문에 data도 cuda에 올려주기
    labels = labels.to(device)

    optimizer.zero_grad()
    outputs = model(inputs)
    _, preds = torch.max(outputs, 1)
    loss = criterion(outputs, labels)
    
    loss.backward()
    optimizer.step()

    running_loss += loss.item() * inputs.size(0)
    running_corrects += torch.sum(preds == labels.data)

  epoch_loss = running_loss / len(train_datasets)
  epoch_acc = running_corrects / len(train_datasets) * 100.

  print('#{} Loss:{:.4f} Acc:{:.4f}% Time:{:.4f}s'.format(epoch,epoch_loss,epoch_acc,time.time()-start_time))
```

```python
# model 저장
torch.save(model.state_dict(), 'model_dict.pth')   # n1
torch.save(model, 'model.pth')    # n2
```

```python
model.eval()
start_time = time.time()

with torch.no_grad():
    running_loss = 0.
    running_corrects = 0

    for inputs, labels in test_dataloader:
        inputs = inputs.to(device)
        labels = labels.to(device)

        outputs = model(inputs)
        _,preds = torch.max(outputs,1)
        loss = criterion(outputs, labels)

        running_loss += loss.item() * inputs.size(0)
        running_corrects += torch.sum(preds == labels.data)

        # 한 배치의 첫 번째 이미지에 대하여 결과 시각화
        print(f'[예측 결과: {class_names[preds[0]]}] (실제 정답: {class_names[labels.data[0]]})')
        imshow(inputs.cpu().data[0], title='예측 결과: ' + class_names[preds[0]])

    epoch_loss = running_loss / len(test_datasets)
    epoch_acc = running_corrects / len(test_datasets) * 100.
    print('[Test Phase] Loss: {:.4f} Acc: {:.4f}% Time: {:.4f}s'.format(epoch_loss, epoch_acc, time.time() - start_time))
```
