```python
import torch

model = torch.hub.load('ultralytics/yolov5','yolov5s')
```

```python
# inference 
from pandas import pandas
url = 'https://img.khan.co.kr/news/2019/11/29/l_2019112901003607500286631.jpg'
result = model(url)
result.pandas().xywh[0]
```
